```python
def average_distance_from_sample(G, n):
    """
    Return the average distance between ``n`` randomly selected
    pairs of points from the graph ``n``.

    INPUT:

    ``G`` - Sage graph.
    ``n`` - positibe integer (sample size)

    OUTPUT:

    floating point real number.
    """
    y=G.vertices()
    from random import sample
    x=sample(range(len(y)),n)
    x1=sample(range(len(y)),n)
    average_distance=0;
    #xx=[]
    for i in range(n):
        average_distance+=G.shortest_path_length(y[x[i]], y[x1[i]])
    average_distance=float(average_distance/n)
    return average_distance
```
